```python
import json
import numpy as np
from tqdm import tqdm
from evaluate import load

wer_metric = load("wer")
```

```python
def read_jsonl(file_path):
    with open(file_path, 'r') as f:
        return [json.loads(line) for line in f]
```

```python
asr_text = read_jsonl("kokoroTTS/chatbot-arena-spoken-1turn-english-whisper-transcript.jsonl")
asr_text = [x['question_transcription'].lower() for x in asr_text]
```

```python
with open("chatbot-arena-spoken-1turn-english.json") as f:
    man_text = json.load(f)
man_text = [x['conversation_a'][0]['content'].lower() for x in man_text]
```

```python
len(man_text)
```

```python
man_text[0]
```

```python
assert len(asr_text) == len(man_text)
```

```python
wer_arr = []
results = []
for asr, man in tqdm(zip(asr_text, man_text)):
    wer = wer_metric.compute(references=[man], predictions=[asr])
    wer = min(1.0, wer)
    wer_arr.append(wer)
    results.append({
        'asr': asr,
        'man': man,
        'wer': wer
    })
```

```python
np.mean(wer_arr)
```

```python
# printing first 100 results with highest WER
results.sort(key=lambda x: x['wer'], reverse=True)
for x in results[:100]:
    print(x['wer'])
    print(x['asr'])
    print(x['man'])
```
