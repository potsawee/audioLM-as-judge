```python
import pandas as pd
import numpy as np
from datasets import load_dataset
```

```python
ds = load_dataset("potsawee/speech-quality-thaimos-pairwise")["train"]
ds[0]
```

```python
# for example, if we look at the second utterance of the first pair
system = ds[0]["system_b"]
file_name = ds[0]["audio_b"]["path"]
mos_score = ds[0]["pronunciation_b"] # average over all workers
print(f"System: {system}, Sentence ID: {file_name}")
print(f"MOS score: {mos_score}")
```

```python
df_raw = pd.read_csv("data_thaimos_master.tsv", sep="\t")
```

```python
# Speech quality here is in the "pronunciation" column (i.e., intelligibility of the synthesized speech)
df_raw[df_raw["system"] == system][df_raw["file_name"] == file_name]
```

```python
# MOS score
df_raw[df_raw["system"] == system][df_raw["file_name"] == file_name]["pronunciation"].mean()
```
