## ChatbotArena-SpokenStyle-SameContent

```python
import re
import os
import json
from glob import glob
from tqdm import tqdm
from datasets import Dataset, Audio
```

```python
def transform_path(path: str) -> str:
    """
    Transforms a path of the form:
        './batch1_generated_gpt_audio/{id}_model_{x}.{style}.wav'
    into:
        './batch1_generated_gpt_audio/counterpart_natural/{id}_model_{x}.natural.wav'
    """
    # Regex breakdown:
    # (./batch1_generated_gpt_audio)  -> capture the leading path
    # /([^_]+)                        -> capture {id} (no underscores)
    # _model_([^\.]+)                 -> capture {x} (anything up to a dot)
    # \.([^\.]+)                      -> capture {style} (anything up to a dot)
    # \.wav$                          -> end with .wav
    pattern = re.compile(r'^(./batch1_generated_gpt_audio)/([^_]+)_model_([^\.]+)\.([^\.]+)\.wav$')
    
    match = pattern.match(path)
    if not match:
        raise ValueError(f"Path '{path}' does not match the expected pattern.")

    base_path, id_, x, style = match.groups()
    
    # Construct the new path
    new_path = f"{base_path}/counterpart_natural/{id_}_model_{x}.natural.wav"
    return new_path
```

```python
with open("../chatbot-arena/chatbot-arena-spoken-1turn-english.json") as f:
    raw_data = json.load(f)
raw_data_dict = {}
for i in range(len(raw_data)):
    raw_data_dict[i] = raw_data[i]
```

```python
# refined question
dir_path = "./refined_questions/"
question_paths = glob(f"{dir_path}*.txt")
refined_question_data = {}
for question_path in question_paths:
    conversation_id = int(question_path.replace(dir_path, "").split(".")[0])
    with open(question_path, 'r') as f:
        refined_question = f.read()
    refined_question = refined_question.strip().strip('"')
    refined_question_data[conversation_id] = refined_question
```

```python
new_data = []
for conversation_id, refined_question in tqdm(refined_question_data.items()):
    wav_path0 = f"./batch1_generated_gpt_audio/{conversation_id}_*.*.wav"
    wav_paths = glob(wav_path0)
    assert len(wav_paths) == 2
    if "natural" in wav_paths[0]:
        wav_path_style = wav_paths[1]
        wav_path_natural = transform_path(wav_path_style)
        style_winner = "model_b"
    elif "natural" in wav_paths[1]:
        wav_path_style = wav_paths[0]
        wav_path_natural = transform_path(wav_path_style)
        style_winner = "model_a"
    else:
        raise Exception()

    if style_winner == "model_a":
        wav_path_model_a = wav_path_style
        wav_path_model_b = wav_path_natural
    else:
        wav_path_model_b = wav_path_style
        wav_path_model_a = wav_path_natural
        
    wav_file = wav_path_style.replace(".wav", "")
    style = wav_file.split(".")[-1]
    raw_data_i = raw_data_dict[conversation_id] 
    refined_question_wav = f"./refined_questions_kokoro_wav/{conversation_id}.wav"

    if style_winner == "model_a":
        assistant = raw_data_i['conversation_a'][1]['content']
    else:
        assistant = raw_data_i['conversation_b'][1]['content']

    assert os.path.exists(refined_question_wav)
    assert os.path.exists(wav_path_model_a)
    assert os.path.exists(wav_path_model_b)
    
    item = {
        'id': conversation_id,
        'original_id': raw_data_i['question_id'],
        'question_refined_wav': refined_question_wav,
        'assistant_a_wav':  wav_path_model_a,
        'assistant_b_wav':  wav_path_model_b,
        'winner_style':  style_winner,
        'style': style,
        'question_refined': refined_question,
        'assistant': assistant,
    }
    new_data.append(item)
```

```python
keys = new_data[0].keys()
# Build a dict of lists
dict_of_lists = {key: [d[key] for d in new_data] for key in keys}
hf_data = Dataset.from_dict(dict_of_lists).cast_column("question_refined_wav", Audio()).cast_column("assistant_a_wav", Audio()).cast_column("assistant_b_wav", Audio())
```
